A line graph is used to represent values that change over time or follow a continuous sequence.

```python
df = pd.read_csv("sales_data.csv")
```

```python
df.head()
```

```python
# Calculate the total quantity of products sold by the company.
total_quantity = df['quantity'].sum()
print(f"QuantitÃ© totale de produits vendus : {total_quantity}")
```

```python
# Identify the category that had the highest revenue and calculate how much revenue it generated.
revenue_by_category = df.groupby('category')['revenue'].sum()
highest_revenue_category = revenue_by_category.idxmax()  # idxmax() = lâ€™index avec la valeur la plus grande(le nom de la catÃ©gorie qui gagne le plus)
highest_revenue_amount = revenue_by_category.max() # category_revenue.max() = la plus grande valeur(la valeur du revenu maximum)

print(f"CatÃ©gorie avec le revenu le plus Ã©levÃ© : {highest_revenue_category}")
print(f"Revenu gÃ©nÃ©rÃ© par cette catÃ©gorie : ${highest_revenue_amount:,.2f}")
```

```python
# 4. DÃ©terminer le revenu moyen par vente
average_revenue_per_sale = df['revenue'].mean()
print(f"Revenu moyen par vente : ${average_revenue_per_sale:,.2f}")
```

```python
# Convertir la colonne 'date' en datetime objects pour extraire les trimestres
df['date'] = pd.to_datetime(df['date'])
df['quarter'] = df['date'].dt.quarter

# Grouper les donnÃ©es par trimestre et sommer les revenus
revenue_by_quarter = df.groupby('quarter')['revenue'].sum()
```

```python
print(revenue_by_quarter)
```

```python
plt.figure(figsize=(7, 5))
plt.bar(revenue_by_quarter.index.astype(str), revenue_by_quarter.values, color='coral')

plt.xlabel("Trimestre (Quarter)")
plt.ylabel("Revenu total ($)")
plt.title("Revenu total gÃ©nÃ©rÃ© par trimestre")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

ðŸŒŸ **Exercise 5 : Data Visualisation Using MatPlotLib**


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_values = np.linspace(-10, 10, 100) # CrÃ©e 100 points entre -10 et 10
y_values = x_values**2 # Calcule y = x^2 pour chaque valeur de x
```

```python
plt.figure(figsize=(6, 4))
plt.plot(x_values, y_values, color='red', linestyle='-', linewidth=2)

# 3. Ajouter les Ã©tiquettes et le titre
plt.title("TracÃ© de la fonction y = x^2")
plt.xlabel("Valeurs de x")
plt.ylabel("Valeurs de y (x^2)")

# 4. Ajouter une grille et afficher le graphique
plt.grid(True, linestyle='--', alpha=0.6)
plt.axhline(0, color='black',linewidth=0.5) # Ligne de l'axe X
plt.axvline(0, color='black',linewidth=0.5) # Ligne de l'axe Y
plt.show()
```

```python

# 1. DonnÃ©es de ventes
products = ['Produit A', 'Produit B', 'Produit C', 'Produit D']
sales = [15, 30, 45, 20]

# 2. CrÃ©er le graphique Ã  barres
plt.figure(figsize=(7, 5))
plt.bar(products, sales, color=['blue', 'green', 'orange', 'red'])

# 3. Ajouter les Ã©tiquettes et le titre
plt.title("Ventes hebdomadaires par produit")
plt.xlabel("Produit")
plt.ylabel("UnitÃ©s vendues")

# 4. Afficher le graphique
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python

# 1. DonnÃ©es
fruits = ['Pommes', 'Bananes', 'Cerises', 'Dattes']
percentages = [40, 30, 20, 10]
colors = ['#ff9999', '#ffe4b5', '#66b2ff', '#99ff99'] # Couleurs personnalisÃ©es

# 2. CrÃ©er le diagramme circulaire
plt.figure(figsize=(6, 6))
# 'autopct' formate l'affichage des pourcentages sur le graphique
plt.pie(percentages, labels=fruits, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)

# 3. Ajouter un titre et une lÃ©gende
plt.title("Fruits prÃ©fÃ©rÃ©s d'un groupe de personnes")
plt.legend(fruits, title="Fruits", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# 4. S'assurer que le cercle est bien rond (equal aspect ratio) et afficher
plt.axis('equal')
plt.show()
```
