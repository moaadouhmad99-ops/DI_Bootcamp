```python
plt.figure(figsize=(8,6))
plt.scatter(df["Social support"], df["Score"], color='blue', alpha=0.6)
plt.title("Relation entre Social Support et Happiness Score (2019)")
plt.xlabel("Social Support")
plt.ylabel("Happiness Score")
plt.grid(True)
plt.show()
```

```python
# Trier par GDP pour visualiser la tendance
df_sorted = df.sort_values("GDP per capita", ascending=False)

fig, ax = plt.subplots(figsize=(30,6))

# Bar plot pour GDP
ax.bar(df_sorted["Country or region"], df_sorted["GDP per capita"], color='skyblue', label="GDP per capita")

# Line plot pour Healthy Life Expectancy
ax.plot(df_sorted["Country or region"], df_sorted["Healthy life expectancy"], marker='o', color='orange', linewidth=2, label="Healthy Life Expectancy")

plt.title("GDP per Capita et Healthy Life Expectancy par pays (2019)")
plt.xlabel("Pays")
plt.ylabel("Valeur")
plt.xticks(rotation=90)
plt.legend()
plt.grid(True)
plt.show()
```
