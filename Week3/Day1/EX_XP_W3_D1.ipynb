# **EXERCISE 1 — Introduction to Data Analysis (Essay)**

**What is data analysis?**

Data analysis is the process of collecting, cleaning, transforming, and interpreting data in order to extract valuable insights and support decision-making. It involves using statistical, mathematical, and computational methods to identify patterns, correlations, and trends within datasets.
In simple terms, it transforms raw data into meaningful information.

**Why is data analysis important in modern contexts?**

Today’s world generates massive amounts of data—from social media, financial systems, healthcare, sensors, and more. Data analysis is essential because it allows organizations, governments, and individuals to make informed decisions based on evidence rather than intuition.
It improves efficiency, predicts future outcomes, identifies problems early, and guides strategic actions.

**Three areas where data analysis is widely applied**

Healthcare:
Used to predict disease outbreaks, personalize treatments, reduce medical errors, and analyze patient records for better decision-making.

Finance and Banking:
Helps detect fraud, evaluate credit risk, optimize investments, and study customer spending behavior.

Business and Marketing:
Companies analyze customer data to understand buying behavior, improve product recommendations, and optimize advertisement targeting.

#**Exercise 5: Basic Data Analysis with Google Colab**

```python
df_mean = df['SepalWidthCm'].mean()
df_median = df['SepalWidthCm'].median()
df_mode = df['SepalWidthCm'].mode()[0] # .mode() returns a Series, take the first if multiple modes exist

print(f"Mean of SepalWidthCm: {df_mean:.2f}")
print(f"Median of SepalWidthCm: {df_median:.2f}")
print(f"Mode of SepalWidthCm: {df_mode:.2f}")
```

```python
import matplotlib.pyplot as plt
plt.plot(df['SepalWidthCm'], df['PetalLengthCm'])
plt.show()
```
