```python
# Load an image using PIL
import matplotlib.pyplot as plt
from PIL import Image
image_path = 'Flower Color Images Dataset/flowers/flowers/19_010.png'
original_image = Image.open(image_path)

# Display the original image using matplotlib
plt.imshow(original_image)
```

```python
from scipy.ndimage import rotate
def rotate_image_30_degrees(image):
  return rotate(image, 30, reshape=False)

rotated_image = rotate_image_30_degrees(original_image)
plt.imshow(rotated_image)
```

```python
#Flip an image horizontally and then vertically using mirror and flip functions from ImageOps
from PIL import ImageOps
flipped_image = ImageOps.mirror(original_image)
plt.imshow(flipped_image)

```

```python
#flip an vertically
flipped_image = ImageOps.flip(original_image)
plt.imshow(flipped_image)
```

```python
from PIL import Image
import matplotlib.pyplot as plt

# Zoom factor
scale = 1.2

# Resize (zoom)
zoomed_image = original_image.resize(
    (int(original_image.width * scale), int(original_image.height * scale)),
    Image.LANCZOS
)

plt.imshow(zoomed_image)
plt.axis('off')
plt.show()

```
