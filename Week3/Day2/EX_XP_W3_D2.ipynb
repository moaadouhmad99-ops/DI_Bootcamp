#Exercise 5 : Export a dataframe to excel format and JSON format.

Créer un simple DataFrame.

Exporter le DataFrame en fichier Excel.

Exporter le DataFrame en fichier JSON.


```python
# Créer un simple DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [24, 30, 29],
    "City": ["Rabat", "Casablanca", "Tanger"]
}

dy = pd.DataFrame(data)

# Exporter le DataFrame en fichier Excel
#pip install openpyxl
dy.to_excel("data.xlsx", index=False)

# Exporter le DataFrame en fichier JSON
dy.to_json("data.json", orient="records", indent=4)

print("Export terminé !")
```
