Data Handling and Analysis in Python

```python
import pandas as pd
df = pd.read_csv("datascience_salaries.csv")
print(df.head())
```

```python
# Normalize the 'salary' column using Min-Max normalization
print(df['salary'].min())
print(df['salary'].max())
df['salary_normalized'] = (df['salary'] - df['salary'].min()) / (df['salary'].max() - df['salary'].min())
print("\nDataFrame after salary normalization:")
print(df.head())
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
numeric_df = df.select_dtypes(include=["int64", "float64"])
numeric_df = numeric_df.dropna()
```

```python
pca = PCA(n_components=2)
pca_components = pca.fit_transform(numeric_df)

df_pca = pd.DataFrame(pca_components, columns=["PCA1", "PCA2"])
df_pca.head()
```

```python
tsne = TSNE(n_components=2, learning_rate="auto", init='random')
tsne_components = tsne.fit_transform(numeric_df)

df_tsne = pd.DataFrame(tsne_components, columns=["TSNE1", "TSNE2"])
df_tsne.head()
```
