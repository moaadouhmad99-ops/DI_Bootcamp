# **Exercise 1: Bar Graph**

**Objective:** Create a bar graph representing the sales data of a retail company.

# **Exercise 2: Hierarchical Indexing**

**Objective:** Filter and retrieve all days when the temperature exceeded 30°C (86°F) for cities in Canada using hierarchical indexing.

```python
import pandas as pd

# Sample data including temperatures for various cities and dates
data = {
    'Country': ['USA', 'Canada', 'USA', 'Canada', 'USA', 'Canada', 'USA', 'Canada', 'Canada', 'USA'],
    'City': ['New York', 'Toronto', 'Los Angeles', 'Vancouver', 'Chicago', 'Montreal', 'Houston', 'Calgary', 'Edmonton', 'Miami'],
    'Date': pd.to_datetime(['2023-07-01', '2023-07-01', '2023-07-01', '2023-07-02', '2023-07-02', '2023-07-02', '2023-07-03', '2023-07-03', '2023-07-04', '2023-07-04']),
    'Temperature': [28, 31, 35, 29, 27, 32, 30, 33, 28, 36] # Temperatures in Celsius
}

df = pd.DataFrame(data)

# Set hierarchical index on ‘Country’, ‘City’, and ‘Date’
df_indexed = df.set_index(['Country', 'City', 'Date'])

print("Original DataFrame with Hierarchical Index:")
display(df_indexed)
```

```python
# Filter: only Canada
df_canada = df_indexed.loc["Canada"]

# Filter: temperature > 30°C
hot_days_canada = df_canada[df_canada["Temperature"] > 30]
print("\nDays in Canada where temperature exceeded 30°C:")
display(hot_days_canada)
```

# **Exercise 4: Visualizing Movie Durations**

**Objective:** Create a distribution plot to visualize the durations of MCU movies.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    "Movie": [
        "Iron Man", "Thor", "Captain America", "Avengers",
        "Guardians of the Galaxy", "Black Panther", "Doctor Strange"
    ],
    "Duration": [126, 115, 124, 143, 121, 134, 115]  # en minutes
})

print(df)
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Duration"], kde=True, bins=8)

plt.title("Distribution of MCU Movie Durations")
plt.xlabel("Duration (minutes)")
plt.ylabel("Frequency")

plt.show()
```

# **Exercise 5: Box Plot of Audience and Tomato Meter Scores**

1. Création d’un DataFrame d’exemple

```python
import pandas as pd

df = pd.DataFrame({
    "Tomato_Meter": [85, 92, 76, 55, 40, 98, 67, 73],
    "Audience_Score": [80, 95, 70, 60, 50, 99, 65, 75]
})
```

2. Melt → passer au format long (long format)

```python
df_melted = df.melt(value_vars=["Tomato_Meter", "Audience_Score"],
                    var_name="Score_Type",
                    value_name="Score")
df_melted
```

3. Boxplot avec Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(7,5))

sns.boxplot(data=df_melted, x="Score_Type", y="Score", palette="Set2")

plt.title("Comparison of Tomato Meter and Audience Scores")
plt.xlabel("Score Type")
plt.ylabel("Score (%)")

plt.show()
```

# **Exercise 6: Pair Plot of Financial Data**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Exemple de dataset MCU (fictif mais réaliste)

```python
data = {
    "Movie": [
        "Iron Man", "Thor", "Avengers", "Black Panther", "Doctor Strange",
        "Captain Marvel", "Ant-Man"
    ],
    "Production_Budget": ["$140M", "$150M", "$220M", "$200M", "$165M", "$175M", "$130M"],
    "Opening_Weekend": ["$98M", "$65M", "$207M", "$202M", "$85M", "$153M", "$57M"],
    "Domestic_Box_Office": ["$318M", "$181M", "$623M", "$700M", "$232M", "$427M", "$180M"],
    "Worldwide_Box_Office": ["$585M", "$449M", "$1.51B", "$1.34B", "$677M", "$1.13B", "$519M"],
    "MCU_Phase": ["Phase 1", "Phase 1", "Phase 1", "Phase 3", "Phase 3", "Phase 3", "Phase 2"]
}

df = pd.DataFrame(data)
```

Convertir les données financières en numérique

```python
def parse_money(x):
    x = x.replace("$", "").replace(",", "")
    if "B" in x:
        return float(x.replace("B", "")) * 1_000_000_000
    if "M" in x:
        return float(x.replace("M", "")) * 1_000_000
    return float(x)
```

```python
cols = ["Production_Budget", "Opening_Weekend", "Domestic_Box_Office", "Worldwide_Box_Office"]

for col in cols:
    df[col] = df[col].apply(parse_money)
```

Créer le Pairplot

```python
sns.pairplot(
    df,
    vars=["Production_Budget", "Opening_Weekend", "Domestic_Box_Office", "Worldwide_Box_Office"],
    hue="MCU_Phase",
    palette="Set2",
    diag_kind="kde"   # KDE diagonal
)

plt.suptitle("Pair Plot of MCU Financial Data", y=1.02)
plt.show()
```
