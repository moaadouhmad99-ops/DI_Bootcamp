# **Exercise 1 : Maximum Value Replacement in Matrix**

```python
import numpy as np

# Create a random 5x5 matrix
A = np.random.rand(5, 5)
print("Original matrix:\n", A)

# Find the index of the maximum value
max_index = np.unravel_index(np.argmax(A), A.shape)

# Replace the maximum value with 0
A[max_index] = 0

print("\nMatrix after replacing max value with 0:\n", A)
```

# **Exercise 2 : Finding Common Values in Arrays**

```python
import numpy as np

# Create two random arrays of size 5
A = np.random.rand(5)
B = np.random.rand(5)
# Round values to 2 decimals
A = np.round(A, 2)
B = np.round(B, 2)

print("Array A:", A)
print("Array B:", B)

# Find common values (intersection)
common_values = np.intersect1d(A, B)

print("\nCommon values:\n", common_values)
```

# **Exercise 3 : Sorting Array in Ascending and Descending Order**

```python
import numpy as np

# Create a random 1D array of size 10
A = np.random.rand(10)
A = np.round(A, 2)
print("Original array:", A)

# Sort in ascending order
A_ascending = np.sort(A)
print("Ascending order:",A_ascending)

# Sort in descending order
A_descending = np.sort(A)[::-1]
print("Descending order:",A_descending)
```

# **Exercise 4 : Matrix Rank Determination**

```python
matric_A = np.random.rand(4,4)
rank_A = np.linalg.matrix_rank(matric_A)
print("Matrix A:\n", matric_A)
print("\nRank of Matrix A:", rank_A)
```

# **Exercise 5 : Creating Border Array**

```python
import numpy as np

# Créer un tableau de zéros
array = np.zeros((5, 5), dtype=int)

# Mettre 1 sur les bords
array[0, :] = 1       # première ligne
array[-1, :] = 1      # dernière ligne
array[:, 0] = 1       # première colonne
array[:, -1] = 1      # dernière colonne


print(array)
```
