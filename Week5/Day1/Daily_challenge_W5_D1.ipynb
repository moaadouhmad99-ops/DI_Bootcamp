**Senario :** you are a data analyst working with a dataset of global weather. Your task is to analyze temperature trends and visualize the results.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Task 1. Data Preparation:
# --- Generate synthetic temperature data ---
temps = np.random.uniform(low=-5, high=35, size=(10, 12))

# City names
cities = [f"City_{i+1}" for i in range(10)]

# Month names
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# --- Create DataFrame ---
df = pd.DataFrame(temps, index=cities, columns=months)

df.head()
```

```python
#task 2. Data Analysis:
# Annual average temperature per city
annual_averge = df.mean(axis=1)

# Identify hottest & coldest cities
hottest_city = annual_averge.idxmax()
coldest_city = annual_averge.idxmin()

annual_averge, hottest_city, coldest_city
```

```python
#task 3. Data Visualization:
plt.figure(figsize=(14, 6))
for city in df.index:
    if city == hottest_city:
        plt.plot(df.columns, df.loc[city], marker="o", linewidth=3, label=f"{city} (Hottest)",color ='red',alpha=0.2)
    elif city == coldest_city:
        plt.plot(df.columns, df.loc[city], marker="s", linewidth=3, label=f"{city} (Coldest)",color ='blue',alpha=0.2)
    else:
        plt.plot(df.columns, df.loc[city], alpha=0.4, label = city)
plt.title("Monthly Temperature Trends for Each City")
plt.xlabel("Month")
plt.ylabel("Temperature (Â°C)")
plt.legend(loc="lower left")
#plt.tight_layout()
plt.show()
```
