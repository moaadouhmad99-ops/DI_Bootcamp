# **Exercise 1 : Minimum and Maximum of Random Array**

```python
import numpy as np

# Create a 5x5 array with random values
arr = np.random.rand(5,5)

# Find minimum and maximum values
minimum_value = arr.min()
maximum_value = arr.max()

print("Random 5x5 array:\n", arr)
print("\nMinimum value:", minimum_value)
print("Maximum value:", maximum_value)
```

# **Exercise 2 : Matrix Normalization**

```python
import numpy as np

# Create a 3x3 random matrix
A = np.random.rand(3, 3)

# Compute mean and standard deviation
mean_A = A.mean()
std_A = A.std()

# Normalize the matrix
A_normalized = (A - mean_A) / std_A

print("Original matrix:\n", A)
print("\nMean:", mean_A)
print("Standard deviation:", std_A)
print("\nNormalized matrix:\n", A_normalized)
```

# **Exercise 4 : Matrix Multiplication**

```python
import numpy as np
A = np.random.rand(5, 3)
B = np.random.rand(3, 2)
C = np.dot(A, B)
print("A=",A)
print("B=",B)
print("A*B=",C)
```

# **Exercise 5**

```python
import numpy as np

# Creating two matrices
A = np.array(np.random.rand(4,3))
B = np.array(np.random.rand(3,4))

# Performing matrix multiplication
result = A @ B  # Equivalent to np.dot(A, B)

print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Matrix Multiplication Result:\n", result)
```
